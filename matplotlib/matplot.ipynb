```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2D line plot

```python
# 2D line plot
# numerical data - numerical data
# numerical data - catogorical data
# use case = Time series data


price=[48000,54000,57000,49000,47000,45000]   # y axis (numerical)
year = [2015,2016,2017,2018,2019,2020]        # x axis (catagorical)

plt.plot(year,price)
```

```python
df = pd.read_csv("sharma-kohli.csv")

plt.plot(df['index'],df['V Kohli'])
```

```python
plt.plot(df['index'],df['RG Sharma'])
```

```python
# plotting multiple plots

plt.plot(df['index'],df['V Kohli'],color='black',linestyle='solid',marker='o',label='Virat')
plt.plot(df['index'],df['RG Sharma'],color='blue',linestyle='solid',marker='o',label='Rohit')
plt.title('King Vs Hitman')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend() # to print labels for diff colors for EX. black is for virat.
```

```python
price=[48000,54000,57000,49000,47000,45000,4500000]   # y axis (numerical)
year = [2015,2016,2017,2018,2019,2020,2021]           # x axis (catagorical)

plt.plot(year,price)

# here the rest of the graph except that 4500000, it looks like flat 
# because the rise is sudden and other values are very low in comparision.

plt.ylim(0,100000)
# here we can see that there is a sudden rise but we also can notice 
# that rest of the graph is not flat because of ylim function
# here we can see variation very clearly

# we also can chane the range of x axis for EX. 2017-2019 by xlim function


# we can add grid here
plt.grid()

# if we use pycharm or we use flask then the above code 
# will not run for running this code we have to use plt.show() function
plt.show()
```

. y axis = aggregate operation perform

```python
x = ['blue','pink','red','green']
y = [10,25,18,8]
plt.bar(x,y,color='lightgreen')
```

```python
data = pd.read_csv("batsman_season_record.csv")
```

```python
data.head()
```

```python
plt.bar(np.arange(data.shape[0])-0.25,data['2015'],width=0.25)
plt.bar(np.arange(data.shape[0]),data['2016'],width=0.25)
plt.bar(np.arange(data.shape[0])+0.25,data['2017'],width=0.25)


plt.xticks(np.arange(data.shape[0]),data['batsman'],rotation='vertical')  # to show names
plt.show()
```

stacked bar chart

```python
plt.bar(data['batsman'],data['2015'],label='2015')
plt.bar(data['batsman'],data['2016'],bottom=data['2015'],label='2016')
plt.bar(data['batsman'],data['2017'],bottom=(data['2015']+data['2016']),label='2017')
plt.legend()
plt.show()
```
